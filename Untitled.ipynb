```python
def normalize(n):
    n /= np.max(np.abs(n))
    n *= (255.0 / n.max())
    return n

def HSV(H, S, V):
    return np.dstack(map(np.array, [normalize(H), normalize(S),
                      normalize(V)])).astype('uint8')
```
